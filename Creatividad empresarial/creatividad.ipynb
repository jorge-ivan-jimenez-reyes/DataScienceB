```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Crear datos dummy
np.random.seed(42)
df = pd.DataFrame({
    'Video_ID': [f'Video_{i}' for i in range(1, 21)],
    'Views': np.random.randint(1000, 10000, 20),
    'Shares': np.random.randint(100, 1000, 20),
    'Followers_Gained': np.random.randint(10, 300, 20)
})

df.head()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Views', y='Shares', data=df)
plt.title('Views vs Shares')
plt.xlabel('Views')
plt.ylabel('Shares')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Views', y='Followers_Gained', data=df, color='green')
plt.title('Views vs Followers Gained')
plt.xlabel('Views')
plt.ylabel('Followers Gained')
plt.show()
```

```python
df_sorted = df.sort_values('Views', ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x='Views', y='Video_ID', data=df_sorted, palette='viridis')
plt.title('Views per Video')
plt.xlabel('Views')
plt.ylabel('Video ID')
plt.show()
```

```python
# Engagement Rate = (Shares + Followers) / Views
df['EngagementRate'] = (df['Shares'] + df['Followers_Gained']) / df['Views']

plt.figure(figsize=(10,6))
sns.barplot(x='EngagementRate', y='Video_ID', data=df.sort_values('EngagementRate', ascending=False), palette='coolwarm')
plt.title('Engagement Rate per Video')
plt.xlabel('Engagement Rate')
plt.ylabel('Video ID')
plt.show()
```

```python
plt.figure(figsize=(12,6))

# Likes
plt.plot(df['Date'], df['Likes'], label='Likes', marker='o', color='blue')

# Videos publicados
plt.plot(df['Date'], df['VideosPosted'] * 400, label='Videos Posted (scaled)', marker='s', color='green', linestyle='--')

plt.title('Tendencias de Likes y Videos Publicados en el Tiempo')
plt.xlabel('Fecha')
plt.ylabel('Cantidad')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
def obtener_ingreso(nombre_concepto):
    while True:
        try:
            valor = float(input(f"Ingrese el monto de ingresos por {nombre_concepto} (en pesos): ") or 0)
            if valor < 0:
                print("El ingreso no puede ser negativo. Intenta de nuevo.")
                continue
            return valor
        except ValueError:
            print("Por favor ingresa un nmero vlido.")

def calcular_ingresos_distintos_salarios():
    print("=== Clculo de Ingresos Distintos a Sueldos y Salarios ===\n")

    arrendamiento = obtener_ingreso("arrendamiento")
    dividendos = obtener_ingreso("dividendos")
    intereses = obtener_ingreso("intereses bancarios")
    actividad_empresarial = obtener_ingreso("actividad empresarial o profesional")
    otros = obtener_ingreso("otros ingresos (premios, donativos, etc.)")

    total = arrendamiento + dividendos + intereses + actividad_empresarial + otros

    print("\n--- Resumen ---")
    print(f"Ingresos por arrendamiento:        ${arrendamiento:,.2f}")
    print(f"Ingresos por dividendos:           ${dividendos:,.2f}")
    print(f"Ingresos por intereses:            ${intereses:,.2f}")
    print(f"Ingresos por actividad empresarial:${actividad_empresarial:,.2f}")
    print(f"Otros ingresos:                    ${otros:,.2f}")
    print(f"\n Total de ingresos distintos a sueldos y salarios: ${total:,.2f}")

if __name__ == "__main__":
    calcular_ingresos_distintos_salarios()
```
