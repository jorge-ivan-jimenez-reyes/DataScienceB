```python
data = {
   "iPhone": "$70B",
   "Mac": "$8B",
   "iPad": "$7B",
   "Watch, AirPods, Home": "$12B",
   "Products": "$96B",
   "Services": "$23B",
   "Revenue": "$120B",
   "Gross profit": "$55B",
   "Cost of revenue": "$65B",
   "Operating profit": "$40B",
   "Operating expenses": "$14B",
   "Product costs": "$58B",
   "Service costs": "$6B",
   "Net profit": "$34B",
   "R&D": "$8B",
   "SG&A": "$7B",
   "Tax": "$6B",
   "Other": "$50M"
}
data.values()
```

<img src="https://developer.mozilla.org/en-US/blog/learn-css-hues-colors-hsl/color-wheel.svg" width=400>

```python
# RGBA https://www.w3schools.com/css/css_colors_rgb.asp
# HSL https://www.w3schools.com/colors/colors_hsl.asp
# Hue, Saturation, Light
colors=["hsl(0, 0%, 50%)" for n in range(7)] # Gris
colors+=["hsl(120, 75%, 50%)"]               # Verde
colors+=["hsl(0, 75%, 50%)"]                 # Rojo
colors+=["hsl(120, 75%, 50%)"]               # Verde
colors+=["hsl(0, 75%, 50%)"]*3               # Rojo
colors+=["hsl(120, 75%, 50%)"]               # Verde
colors+=["hsl(0, 75%, 50%)"]*4               # Rojo
```

```python
nodes={"label":list(data.keys()),
       "customdata":list(data.values()),
       "color":colors}
```

```python
links=[
 {"source":0, "target":4, "value":70},
 {"source":1, "target":4, "value":8},
 {"source":2, "target":4, "value":7},
 {"source":3, "target":4, "value":12},
 {"source":4, "target":6, "value":96},
 {"source":5, "target":6, "value":23},
 {"source":6, "target":7, "value":55},
 {"source":6, "target":8, "value":65},
 {"source":7, "target":9, "value":40},
 {"source":7, "target":10, "value":14},
 {"source":8, "target":11, "value":58},
 {"source":8, "target":12, "value":6},
 {"source":9, "target":13, "value":34},
 {"source":9, "target":14, "value":6},
 {"source":9, "target":15, "value":.05},
 {"source":10, "target":16, "value": 8},
 {"source":10, "target":17, "value": 7},
]
```

```python
links[0]['source']
```

```python
import plotly.graph_objects as go
sankey=go.Sankey(
    node=nodes,
    link={
        "source":[s["source"] for s in links],
        "target":[t["target"] for t in links],
        "value":[v["value"] for v in links],
        "color":[c.replace("50","75") for c in colors]
    })
```

```python
fig= go.Figure(sankey)
fig.show()
```
